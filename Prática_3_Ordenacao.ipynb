**1. Método Rápido - QuickSort**

O método QuickSort se baseia na estratégia de **dividir para conquistar**. No início, seleciona-se um valor, chamado de pivô.

Embora existam muitas maneiras de selecionar o pivô, iremos utilizar o primeiro item da lista.

O papel do pivô é ajudar na divisão dos elementos da lista em duas partes: **menores** que o pivô e **maiores**. A posição à qual o pivô pertence de fato na lista ordenada, conhecida como ponto de divisão, será usada para quebrar a lista nas duas partes (menores e maiores). Cada uma das partes deve recomeçar o método QuickSort.

Veja mais detalhes e exemplo em: https://panda.ime.usp.br/panda/static/pythonds_pt/05-OrdenacaoBusca/OQuickSort.html

Faça testes e experimentos no QuickSort em: https://visualgo.net/en/sorting

````python
def particiona(V, inicio, final):
    esq = inicio
    dir = final
    md = int((esq+dir)/2)

    temp = V[inicio]
    V[inicio] = V[md]
    V[md] = temp

    pivo = V[inicio]
                         # indicação do pivô: elemento do início
    while(esq < dir):
        while(esq <= final and V[esq] <= pivo):
            esq = esq + 1

        while(dir >= 0 and V[dir] > pivo):
            dir = dir - 1

        if(esq < dir):
            aux = V[esq]
            V[esq] = V[dir]
            V[dir] = aux

    V[inicio] = V[dir]
    V[dir] = pivo
    return dir

def quickSort(V, inicio, fim):
    if(fim > inicio):
        pivo = particiona(V, inicio, fim)  # particiona a lista em menores que o pivô e maiores
        quickSort(V, inicio, pivo-1)       # recomeça a ordenação para os menores
        quickSort(V, pivo+1, fim)          # recomeça a ordenação para os maiores
````

````python
lista = [16,13,28,1,9,6,10,24,36,11,5,2]
quickSort(lista,0,11)
print(lista)
````

**2. Revisão em Geração de Listas:**

O trecho de código abaixo mostra como podemos criar funções para gerar listas com valores aleatórios, valores em ordem crescente, valores em ordem decrescente, e valores com muitas repetições.

Estas funções podem ser utilizadas para analisar como o método de ordenação se comporta considerando diferentes tipos de entradas. Vamos testar o método de ordenação rápida (`QuickSort`) nas diferentes situações e comparar os tempos.

````python
from random import randint

def gerarListaValoresAleatorios(quantidade):
   resposta = []
   for i in range(quantidade):
       resposta.append(randint(0,1000))  # valor aleatório entre 0 e 1000
   return resposta

def gerarListaOrdenada(quantidade):
   resposta = []
   inicio = randint(0,1000)
   for i in range(quantidade):
       resposta.append(inicio)
       inicio = inicio + 3
   return resposta

def gerarListaInvertida(quantidade):
   resposta = []
   inicio = randint(0,1000)
   for i in range(quantidade):
       resposta.append(inicio)
       inicio = inicio - 3
   return resposta

def gerarListaPoucasChaves(quantidade, max_chaves):  #max_chaves representa o número de valores distintos
   resposta = []
   for i in range(quantidade):
       resposta.append(randint(1,max_chaves))
   return resposta
````

**3. Função para avaliar o método QuickSort quando aplicado à entradas diferentes**

A função `ordenaQuickSortDiferentesListas` implementa as seguintes ações:
1. Cria quatro listas diferentes (aleatória, ordenada, invertida, com poucas chaves)
2. Ordena cada uma das listas e marca o tempo
3. Imprime os tempos de cada ordenação (mantendo a sequência)

````python
import time

def ordenaQuickSortDiferentesListas(max):

    l1 = gerarListaValoresAleatorios(max)
    l2 = gerarListaOrdenada(max)
    l3 = gerarListaInvertida(max)
    l4 = gerarListaPoucasChaves(max,8)

    ini1 = time.time()
    quickSort(l1,0,max-1)
    fim1 = time.time()
    taleatorio= fim1-ini1

    ini2 = time.time()
    quickSort(l2,0,max-1)
    fim2 = time.time()
    tordenado = fim2-ini2

    ini3 = time.time()
    quickSort(l3,0,max-1)
    fim3 = time.time()
    tinvertido = fim3-ini3

    ini4 = time.time()
    quickSort(l4,0,max-1)
    fim4 = time.time()
    tpoucaschaves = fim4-ini4

    listatempos = [taleatorio, tordenado, tinvertido, tpoucaschaves]

    print(listatempos)

ordenaQuickSortDiferentesListas(1000)
````
